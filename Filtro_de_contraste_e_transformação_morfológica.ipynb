```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show_image(image, title='Image', cmap_type='gray'):
  plt.imshow(image, cmap=cmap_type)
  plt.title(title)
  plt.axis('off')
```

```python
soaps_image = plt.imread('sabonetes_deco.jpg')
```

```python
from skimage import color
from skimage.filters import sobel
```

```python
#Imagem em escala de cinza
soaps_image_gray = color.rgb2gray(soaps_image)
```

```python
#Aplicando detecção de filtro de bordas
edge_sobel = sobel(soaps_image_gray)
```

```python
#Mostrando a imagem original
show_image(soaps_image, 'Original')
```

```python
#Mostrabdo o resultado da imagem para a comparação
show_image(edge_sobel, 'Edge with Sobel')
```

```python
from skimage.filters import gaussian
building_image = plt.imread('Einstein.jpg')
```

```python
#Aplicando filtro gaussiano
gaussian_image = gaussian(building_image, multichannel=True)
```

```python
#Mostrando a imagem original
show_image(building_image, 'Original')
```

```python
#Removendo ruídos
show_image(gaussian_image, 'Redução de ruídos Gaussian')
```
